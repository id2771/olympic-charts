```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
greece = df[df['NOC'] == 'GRE']
greece
```

```python
greece_medals = greece[greece['Medal'].notna()]
greece_medals
```

```python
greece_summer = greece_medals[greece_medals['Season'] == 'Summer']
greece_summer
```

```python
greece_editions = greece_summer.groupby('Year').size()
greece_editions
```

```python
plt.figure(figsize=(12,7))
greece_editions.plot(kind = 'bar')
plt.title("Nombre de médailles de la Grèce aux JOs d'été")
plt.xlabel("Année")
plt.ylabel("Nombre de médailles")
plt.show()
```

```python
greece_editions_modern = greece_summer[greece_summer['Year'] >= 1908]
greece_editions_modern
```

```python
greece_2 = greece_editions_modern.groupby('Year').size()
greece_2
```

```python
plt.figure(figsize=(12,7))
greece_2.plot(kind = 'bar')
```
