```python
import pandas as pd

# Charger les jeux de données
athlete_df = pd.read_csv('athlete_events.csv')
noc_df = pd.read_csv('noc_regions.csv')
```

```python
# Tous les codes NOC
all_nocs = set(noc_df['NOC'])
```

```python
# Codes NOC ayant au moins une médaille (colonne 'Medal' non nulle)
medal_nocs = set(athlete_df.loc[athlete_df['Medal'].notna(), 'NOC'])
```

```python
# Codes NOC sans aucune médaille
no_medal_nocs = all_nocs - medal_nocs
```

```python
# Filtrer la liste des NOC sans médailles et afficher leur nom de région
no_medal_df = noc_df[noc_df['NOC'].isin(no_medal_nocs)][['NOC', 'region', 'notes']].reset_index(drop=True)
```

```python
# Afficher le résultat à l'utilisateur
import ace_tools as tools; tools.display_dataframe_to_user("Pays sans médailles olympiques", no_medal_df)
```

```python
output_path = 'noc_sans_medaille.csv'
no_medal_df.to_csv(output_path, index=False,encoding='utf-8-sig')
print(f"Liste des pays/régions sans médaille enregistrée dans : {output_path}\n")
```
