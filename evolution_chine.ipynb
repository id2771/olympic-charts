```python
# importer les packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# charger les données
df = pd.read_csv('athlete_events.csv')
```

```python
# on garde les lignes qui concernent la Chine
# on fait deux tableaux : un pour les éditions d'été (Season = Summer)
# et un tableau pour l'hiver (Season = Winter)


df_chine = df[df['NOC'] == 'CHN']
df_chine
```

```python
# on garde les lignes où la Chine a gagné une médaille
china_medals = df_chine[df_chine['Medal'].notna()]
```

```python
# on fait deux tableaux : Jeux d'Été / Jeux d'Hiver
df_summer = china_medals[china_medals['Season'] == 'Summer']
df_summer
```

```python
# on regroupe par année et par type de médaille
medailles_par_annee = df_summer.groupby(['Year', 'Medal']).size().unstack(fill_value=0)
```

```python
medailles_par_annee.plot(kind='bar', figsize=(7,12))
plt.title("Nombre de médailles de la Chine par édition (jeux d'été)")
plt.xlabel('Year')
plt.ylabel('Nb de médailles')
plt.legend(title='Type de médaille')
plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```
