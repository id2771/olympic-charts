```python
# importer les packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# charger les données
df = pd.read_csv('athlete_events.csv')
```

```python
# on garde les lignes qui concernent la Chine
df_chine = df[df['NOC'] == 'CHN']
df_chine
```

```python
# à l'étape d'avant, ajouter cette condition pour prendre que les versions été
df['Season'] == 'Summer'
```

```python
# on garde les lignes où la Chine a gagné une médaille
china_medals = df_chine[df_chine['Medal'].notna()]
```

```python
# on regroupe par année et par type de médaille
medailles_par_annee = china_medals.groupby(['Year', 'Medal']).size().unstack(fill_value=0)
```

```python
medailles_par_annee.plot(kind='bar', figsize=(7,12))
plt.title('Nombre de médailles de la Chine par édition')
plt.xlabel('Year')
plt.ylabel('Nb de médailles')
plt.legend(title='Type de médaille')
plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```
