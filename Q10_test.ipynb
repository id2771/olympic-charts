```python
# liste de tous les pays
all_nocs = set(noc_regions['NOC'])
```

```python
# liste des pays ayant gagné au moins une médaille
winners = set(df.loc[df['Medal'].notna(), 'NOC'])
```

```python
# liste des pays n'ayant gagné aucune médaille
losers = all_nocs - winners
```

on modifie noc_regions pour ne garder que les pays qui ont gagné des médailles

```python
# les pays ayant gagné des médailles

medal_countries = df.loc[df['Medal'].notna(), 'NOC'].unique()
```

```python
# dataframe noc_regions mais avec que les pays vainqueurs 

noc_medals = noc_regions[noc_regions['NOC'].isin(medal_nocs)]
```

```python
import pandas as pd

# supposons que vous ayez déjà :
#   world      = gpd.read_file(url)
#   medal_nocs = liste ou array de vos NOC qui ont gagné au moins une médaille

# 1. Construire la liste des colonnes de world à tester
candidates = [
    col for col in world.columns
    if col.upper().endswith(('_A2', '_A3', '_N3'))
       or col.upper().startswith(('WB_A2','WB_A3'))
       or 'SOV_' in col.upper()
       or 'ADM0_' in col.upper()
]

# 2. Pour chaque colonne, compter combien de valeurs figurent dans medal_nocs
matches = {
    col: world[col].isin(medal_nocs).sum()
    for col in candidates
}

# 3. Transformer en Series et trier pour voir le classement
best = pd.Series(matches).sort_values(ascending=False)
print(best)
```
